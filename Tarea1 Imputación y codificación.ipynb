### 2. CODIFICACIÓN DE VARIABLES CATEGORICAS

```python
def getCategoryVars(dataset):
    cat_colnames =[]
    colnames=dataHP_impfinal.columns
    for col in colnames:
        if(dataHP_impfinal[col].dtype=='object'):
           cat_colnames.append(col)
    return cat_colnames
```

```python
cat_cols = getCategoryVars(dataHP_impfinal)
cat_cols
```

```python
### Se identificaron 34 columnas con variables categoricas, las cuales hay que codificar

dataset_categoricas = dataHP_impfinal[cat_cols]
dataset_categoricas.head()
```

#### Se realizará codificación a través de One Hot Encoding

```python
categoricas_cod = pd.get_dummies(dataset_categoricas)
categoricas_cod.head()
```

```python
dataHP_sinCategoricas = dataHP_impfinal.drop(dataset_categoricas, axis=1)
dataHP_sinCategoricas.head()
```

```python
DataHP_imp_cod = dataHP_sinCategoricas.join(categoricas_cod, how='inner')
DataHP_imp_cod.head()
```
